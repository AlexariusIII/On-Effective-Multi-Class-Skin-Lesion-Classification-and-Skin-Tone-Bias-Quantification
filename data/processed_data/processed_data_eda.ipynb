```python
import pandas as pd
```

```python
df = pd.read_csv("combined_data.csv")
df
```

```python
for dataset in set(df['dataset'].to_list()):
    print(dataset)
    dataset_df = df[df['dataset']==dataset]
    clinic_df = dataset_df[dataset_df['image_type']=='clinical']
    derma_df = dataset_df[dataset_df['image_type']=='dermatoscopic']
    print(f"Clinic: {len(clinic_df)}")
    print(f"Derma: {len(derma_df)}")
    print()
```
